## 任务一  字典转换

```python
# -*- coding: utf-8 -*-
import pandas as pd

# 新建字典 
df_before = pd.DataFrame()
df_before['编码'] = ['ASD', 'SDF', 'DFG']
df_before['效率'] = ['12', '35', '56']

# 新建字典
dict_test = {'info':[ {'A':1, 'B':'123', 'C': 'ASD'}, 
                     {'A':1, 'B':'234', 'C':'SDF'}, 
                     {'A':1, 'B':'345', 'C':'DFG'}]}

# 值替换
list_get = dict_test['info']
for i in list_get:
    df_before.replace(i['C'], i['B'], inplace=True)
```

## 任务二：条形图

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# 下面两行mac显示中文字体
import matplotlib.pyplot as plt
plt.rcParams["font.family"]="Songti SC"  

# 条形图配置
ig, axes = plt.subplots(2, 2, figsize=(9, 9), dpi=600) # 前两个参数为子图布局， figsize为画布大小，dpi为清晰度设置

# 数据筛选
df_get = pd.read_excel('/Users/yqsl/Desktop/test/data_raw.xlsx')  # 读入数据，需要修改路径

df_get_0 = df_get.query('label==0')
df_get_1 = df_get.query('label==1')
df_get_2 = df_get.query('label==2')


# x轴，y轴， 数据集， ax = 子图分布位置，  ci=0 取消竖线， palette配色方案
sns.barplot(x="商家", y="数量", data=df_get_0, ax=axes[0, 0], palette="Blues_d")
sns.barplot(x="商家", y="数量", data=df_get_1, ax=axes[0, 1], palette="Blues_d")
sns.barplot(x="商家", y="数量", data=df_get_2, ax=axes[1, 0], palette="Blues_d")
```

```python
g = sns.FacetGrid(df_get,col='label',size=6, aspect=0.5, palette="Blues_d")
g.map(sns.barplot,"商家","数量")
```
